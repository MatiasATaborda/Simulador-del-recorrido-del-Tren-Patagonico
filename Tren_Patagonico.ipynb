```python
def mostrar_recorrido(servicio):
    print("\n🚉 Recorrido del tren:")
    for parada in servicio:
        print(f"{parada['estacion']} → {parada['hora']}")

def consultar_tramo(servicio, origen, destino):
    estaciones = [p["estacion"] for p in servicio]

    if origen not in estaciones or destino not in estaciones:
        print("❌ Estación inválida")
        return

    i_origen = estaciones.index(origen)
    i_destino = estaciones.index(destino)

    if i_origen > i_destino:
        print("⚠️ El tren no pasa en ese orden en este servicio")
        return

    tramo = servicio[i_origen:i_destino + 1]

    print("\n🕒 Horarios del tramo:")
    for parada in tramo:
        print(f"{parada['estacion']} → {parada['hora']}")
```
